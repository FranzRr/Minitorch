##Parámetros

Los parámetros son valores que definen las decisiones de un modelo. \\
La página también nos muestra un ejemplo de un modelo lineal y clasificación binaria: \\
$m(x_1,x_2;w_1,w_2,b) = x_1w_1 + x_2w_2 + b$ \\
donde $w_1,w_2,b$ son parámetros y $x_1,x_2$ son el punto de entrada.

##Pérdida
Todo modelo tendrá una función de pérdida que indica que tan bien o mal se está prediciendo. El objetivo que tenemos es minimizar lo más posible esta función de pérdida (Esto dependerá de qué modelo se escoja)

##Ajustando parámetros

Lo que busca el ajuste de parámetros es es encontrar un buen modelo dada una función de pérdida. \\
La biblioteca de minitorch viene con un método de ajuste de parámetros: el descenso de la gradiente. Que es de la siguiente forma: \\
1. Se calcula función pérdida L, con los parámetros.
2. Se puede ver cómo cambiando los parámetros varía la pérdida. (girar las perillas de los parámetros).
3. Actualizar los parámetros con un pequeño cambio en la dirección que localmente reduzca más la pérdida. \\

Pero ir ajustando parámetro por parámetro es ineficiente, por eso se puede tomar un conjunto de derivadas (gradiente) y así sabremos en qué dirección deberán moverse.

##Redes Neuronales


Existen conjuntos de datos que no se pueden separar linealmente. \\
Entonces, un modelo alternativo es una red neuronal. Intuitivamente, las redes neuronales dividen la clasificación en dos o más etapas. Cada etapa utiliza un modelo lineal para remodelar los datos en nuevos puntos. Y la etapa final es un clasificador lineal sobre el punto transformado. \\
Una función para hacer esto es la función ReLU (Eliminar valores por debajo de 0). \\
Por ejemplo, en la imagen las X superiores son positivas y las O y X inferiores son 0.
